````python
# ╔══════════════════════════════════════════════════════════════════════╗
# ║  SECTION 2: One-Click Installation Wizard                           ║
# ║  Automated workspace and lakehouse provisioning                     ║
# ╚══════════════════════════════════════════════════════════════════════╝

import uuid
from datetime import datetime

def run_installation_wizard(config: dict = None):
    """Execute automated installation of Insurance Fabric Accelerator.
    
    Args:
        config: Installation configuration
            - workspace_name: Name for new workspace
            - capacity_id: Fabric capacity ID
            - domain_id: Business domain ID (optional)
            - generate_sample_data: Whether to generate demo data
    """
    print("\n" + "="*70)
    print("🚀 INSURANCE FABRIC ACCELERATOR - INSTALLATION WIZARD")
    print("="*70)
    
    # Default configuration
    if config is None:
        config = {
            "workspace_name": f"Insurance-Accelerator-{datetime.now().strftime('%Y%m%d')}",
            "capacity_id": None,  # User must provide
            "domain_id": None,
            "generate_sample_data": True,
            "deploy_power_bi": True,
            "enable_git": False
        }
    
    installation_id = str(uuid.uuid4())
    steps_completed = []
    
    print(f"\n📋 Installation Configuration:")
    print(f"   Installation ID: {installation_id}")
    print(f"   Workspace Name: {config['workspace_name']}")
    print(f"   Generate Sample Data: {config['generate_sample_data']}")
    print(f"   Deploy Power BI: {config['deploy_power_bi']}")
    
    # === STEP 1: Create Workspace ===
    print("\n" + "-"*70)
    print("STEP 1/7: Creating Fabric Workspace")
    print("-"*70)
    
    api = FabricAPI()
    
    workspace_body = {
        "displayName": config["workspace_name"],
        "description": "Insurance Fabric Accelerator - Automated Installation"
    }
    
    if config.get("capacity_id"):
        workspace_body["capacityId"] = config["capacity_id"]
    
    workspace_response = api.post("/v1/workspaces", workspace_body)
    
    if workspace_response.get("success"):
        workspace_id = workspace_response.get("id")
        print(f"   ✅ Workspace created: {workspace_id}")
        steps_completed.append("workspace_created")
    else:
        print(f"   ❌ Failed to create workspace: {workspace_response.get('error')}")
        return {"status": "failed", "step": "workspace_creation"}
    
    # === STEP 2: Assign to Domain ===
    if config.get("domain_id"):
        print("\n" + "-"*70)
        print("STEP 2/7: Assigning to Business Domain")
        print("-"*70)
        
        domain_response = api.post(
            f"/v1/admin/domains/{config['domain_id']}/assignWorkspaces",
            {"workspacesIds": [workspace_id]}
        )
        
        if domain_response.get("success"):
            print(f"   ✅ Assigned to domain: {config['domain_id']}")
            steps_completed.append("domain_assigned")
        else:
            print(f"   ⚠️  Domain assignment skipped")
    
    # === STEP 3: Create Lakehouses ===
    print("\n" + "-"*70)
    print("STEP 3/7: Creating Medallion Lakehouses")
    print("-"*70)
    
    lakehouses = ["insurance_bronze", "insurance_silver", "insurance_gold", "insurance_metadata"]
    created_lakehouses = {}
    
    for lakehouse_name in lakehouses:
        lakehouse_response = api.post(
            f"/v1/workspaces/{workspace_id}/lakehouses",
            {"displayName": lakehouse_name}
        )
        
        if lakehouse_response.get("success"):
            lakehouse_id = lakehouse_response.get("id")
            created_lakehouses[lakehouse_name] = lakehouse_id
            print(f"   ✅ Created: {lakehouse_name}")
        else:
            print(f"   ❌ Failed: {lakehouse_name}")
    
    steps_completed.append("lakehouses_created")
    
    # === STEP 4: Import Notebooks ===
    print("\n" + "-"*70)
    print("STEP 4/7: Importing Notebooks")
    print("-"*70)
    
    print(f"   ℹ️  14 notebooks ready to import")
    print(f"   ℹ️  Use Fabric UI or REST API to complete notebook import")
    steps_completed.append("notebooks_ready")
    
    # === STEP 5: Generate Sample Data ===
    if config.get("generate_sample_data"):
        print("\n" + "-"*70)
        print("STEP 5/7: Generating Sample Data")
        print("-"*70)
        
        print(f"   ℹ️  Run notebook: 01_demo_data_generator.ipynb")
        print(f"   ℹ️  This will create:")
        print(f"      - 10,000 customers")
        print(f"      - 20,000 policies")
        print(f"      - 5,000 claims (8% fraud)")
        print(f"      - 15,000 invoices")
        print(f"      - 30,000 journal entries")
        steps_completed.append("sample_data_ready")
    
    # === STEP 6: Deploy Power BI Semantic Model ===
    if config.get("deploy_power_bi"):
        print("\n" + "-"*70)
        print("STEP 6/7: Deploying Power BI Semantic Model")
        print("-"*70)
        
        print(f"   ℹ️  Create Direct Lake semantic model")
        print(f"   ℹ️  Source: Gold lakehouse tables")
        print(f"   ℹ️  Use notebook: 90_central_cockpit_dashboard.ipynb for config")
        steps_completed.append("power_bi_ready")
    
    # === STEP 7: Enable Git Integration (Optional) ===
    if config.get("enable_git"):
        print("\n" + "-"*70)
        print("STEP 7/7: Enabling Git Integration")
        print("-"*70)
        
        print(f"   ℹ️  Connect workspace to Git repository")
        print(f"   ℹ️  Use Fabric UI: Workspace Settings → Git Integration")
        steps_completed.append("git_ready")
    
    # === Installation Summary ===
    print("\n" + "="*70)
    print("✅ INSTALLATION COMPLETED")
    print("="*70)
    
    print(f"\n📊 Summary:")
    print(f"   Workspace ID: {workspace_id}")
    print(f"   Lakehouses Created: {len(created_lakehouses)}")
    print(f"   Steps Completed: {len(steps_completed)}")
    
    print(f"\n🚀 Next Steps:")
    print(f"   1. Open workspace: {config['workspace_name']}")
    print(f"   2. Run notebook: 01_demo_data_generator.ipynb")
    print(f"   3. Run notebook: 30_medallion_transformations.ipynb")
    print(f"   4. Open Power BI dashboard to view metrics")
    
    return {
        "status": "success",
        "workspace_id": workspace_id,
        "lakehouses": created_lakehouses,
        "steps_completed": steps_completed
    }

# Example: Run installation
# installation_result = run_installation_wizard({
#     "workspace_name": "Insurance-Accelerator-Demo",
#     "capacity_id": "your-capacity-id",
#     "generate_sample_data": True,
#     "deploy_power_bi": True
# })

print("✅ Installation wizard ready")
````
